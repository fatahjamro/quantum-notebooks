```python
import qiskit
import qiskit.visualization as viz
```

```python
# Create blank circuit with three qubits.
qc = qiskit.QuantumCircuit(3)

# Apply H-gate to each qubit:
for qubit in range(3):
  qc.h(qubit)

# Draw the circuit as it is.
qc.draw(output='mpl')
```

```python
# Create Aer simulator backend.
svsim = qiskit.Aer.get_backend('aer_simulator')

# Save the statevector.
qc.save_statevector()

# Assemble.
qobj = qiskit.assemble(qc)

# Run.
final_state = svsim.run(qobj).result().get_statevector()

# Print statevector.
viz.array_to_latex(final_state)
```

```python
# Print without latex.
print(final_state)
```

```python
# Create a circuit with two qubits.
qc = qiskit.QuantumCircuit(2)

# H gate on qubit 0.
qc.h(0)
# X gate on qubit 1.
qc.x(1)

# Show the circuit.
qc.draw(output='mpl')
```

```python
# Aer simulator.
usim = qiskit.Aer.get_backend('aer_simulator')

# Save unitary.
qc.save_unitary()

# Assemble.
qobj = qiskit.assemble(qc)

# Run.
unitary = usim.run(qobj).result().get_unitary()
```

```python
# See the array.
viz.array_to_latex(unitary)
```

```python
# Circuit with two qubits.
qc = qiskit.QuantumCircuit(2)

# X gate on qubit 1.
qc.x(1)

# Draw circuit.
qc.draw(output='mpl')
```

```python
# Simulate the unitary
usim = qiskit.Aer.get_backend('aer_simulator')

# Circuit as matrix.
qc.save_unitary()

# Assemble.
qobj = qiskit.assemble(qc)

# Run.
unitary = usim.run(qobj).result().get_unitary()

# Display the results:
viz.array_to_latex(unitary)
```

```python
# Two qubit circuit.
qc = qiskit.QuantumCircuit(2)

# CNOT, 1 controlled by 0.
qc.cx(0,1)

# Show the circuit.
qc.draw(output='mpl')
```

```python
# Two qubit circuit.
qc = qiskit.QuantumCircuit(2)

# H gate on 0.
qc.h(0)

# Show circuit.
qc.draw(output='mpl')
```

```python
# Run circuit.
svsim = qiskit.Aer.get_backend('aer_simulator')

# State vector.
qc.save_statevector()

# Assemble.
qobj = qiskit.assemble(qc)

# Final state.
final_state = svsim.run(qobj).result().get_statevector()

# Show statevector.
viz.array_to_latex(final_state)
```

```python
# Two qubits.
qc = qiskit.QuantumCircuit(2)

# H on 0.
qc.h(0)

# CNOT 0 -> 1
qc.cx(0,1)

# Show circuit.
qc.draw(output='mpl')
```

```python
# Initial state.
qc.save_statevector()

# Assemble.
qobj = qiskit.assemble(qc)

# Simulate.
result = svsim.run(qobj).result()

# Final state.
final_state = result.get_statevector()

# Dispaly.
viz.array_to_latex(final_state)
```

```python
# Show counts.
viz.plot_histogram(result.get_counts())
```

```python
# Show Bloch spheres.
viz.plot_bloch_multivector(final_state)
```

```python
# Q Shere.
viz.plot_state_qsphere(final_state)
```
