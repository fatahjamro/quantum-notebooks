```python
# Quantum circuits.
import qiskit

# Plots.
import matplotlib.pyplot as plt

# New figure.
fig, ax = plt.subplots(figsize=(16, 8))

# Create Quantum Circuit - two qubits, two classical bits.
circuit = qiskit.QuantumCircuit(2, 2)

# Hadamard gate on first qubit.
circuit.h(0)

# CNOT gate with control qubit 0 and target qubit 1.
circuit.cx(0, 1)

# Measure the qubits into the classical bits.
circuit.measure([0,1], [0,1])

# Draw the circuit.
circuit.draw(output='mpl', ax=ax)

# Save the figure.
fig.savefig('bell.png', transparent=True)
```

```python
# Simulate.
import qiskit.providers.aer as aer

# Use Aer's qasm simulator.
simulator = aer.QasmSimulator()

# Compile the circuit for the simulator.
compiled_circuit = qiskit.transpile(circuit, simulator)

# Execute circuit on simulator.
job = simulator.run(compiled_circuit, shots=1000)

# Get results.
result = job.result()

# See results.
counts = result.get_counts(compiled_circuit)
print(f"{counts}");
```

```python
# Visualize.
import qiskit.visualization as vis

# Plot histogram.
vis.plot_histogram(counts, figsize=(8, 6))
```
