# Simon's Problem

***

```python
import itertools as it
```

```python
n = 3
```

```python
words = lambda: it.product([0, 1], repeat=n)
```

```python
list(words())
```

```python
guesses = lambda: it.permutations(words())
```

```python
list(guesses())
```

```python
funcs = lambda: it.combinations(words(), 4)
```

```python
list(funcs())
```
